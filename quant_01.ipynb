```python
# 我的第一个A股量化策略：双均线金叉选股
import pandas as pd

def check_signal(stock_code):
    """
    逻辑：当 5日均线 穿过 20日均线（金叉），且属于主流龙头股时，产生买入信号
    """
    print(f"正在分析股票: {stock_code}")
    
    # 这里未来会用 AkShare 抓取真实的K线数据
    ma5 = 15.5
    ma20 = 15.0
    
    if ma5 > ma20:
        return "【买入信号】5日线超越20日线，控制回撤，分批建仓！"
    else:
        return "【观望】继续等待时机"

# 测试策略
print(check_signal("600519")) # 茅台试试
```
